```java
import java.io.BufferedReader;
import java.io.FileReader;
import java.io.IOException;
import java.util.ArrayList;

public class DataSetComponent {
  private String dataSource = null;
  private String[] attributes = null;
  private String[][] instances = null;
  
  public DataSetComponent() {
    /* nothing */
  }

  public String getDataSource() {
    return dataSource;
  }

  public void setDataSource(String dataSource) {
    this.dataSource = dataSource;
    if (dataSource == null) {
      attributes = null;
      instances = null;
    } else
      readDS();
  }
  
  public String[] requestAttributes() {
    return attributes;
  }
  
  public String[][] requestInstances() {
    return instances;
  }
  
  private void readDS() {
    ArrayList<String[]> instArray = new ArrayList<String[]>();
    try {
      BufferedReader file = new BufferedReader(new FileReader(dataSource));
        
      String line = file.readLine();
      if (line != null) {
        attributes = line.split(",");
        line = file.readLine();
        while (line != null) {
          String[] instLine = line.split(",");
          instArray.add(instLine);
          line = file.readLine();
        }
        instances = instArray.toArray(new String[0][]);
      }
        
      file.close();
    } catch (IOException erro) {
      erro.printStackTrace();
    }
  }
  
}
```

```java
import java.util.Random;

public class Gerasints {
    private String[] primeira_linha_matriz;
    private String[][] outras_linhas;
    private String[] doencagerada;
    private String[] saida;
    private String[] doencas;
    
    public Gerasints(String arq){
        DataSetComponent aux = new DataSetComponent();
        aux.setDataSource(arq);
        primeira_linha_matriz = aux.requestAttributes();
        outras_linhas = aux.requestInstances();
    }
    public String[] doenca_sintomas(){
        Random random = new Random();
        int numero = random.nextInt(outras_linhas.length);
        String[] doencagerada;
        doencagerada = outras_linhas[numero];
        int auxi = 0;
        for(int j=0;j<doencagerada.length;j++){
            if (doencagerada[j].equalsIgnoreCase("t") || doencagerada[j].equalsIgnoreCase("true") || doencagerada[j].equalsIgnoreCase("1")){
                auxi++;
            }
        }
        saida = new String[auxi];
        auxi = 0;
        for(int j=0;j<doencagerada.length;j++){
            if (doencagerada[j].equalsIgnoreCase("t") || doencagerada[j].equalsIgnoreCase("true") || doencagerada[j].equalsIgnoreCase("1")){
                saida[auxi] = primeira_linha_matriz[j] ;
                auxi++;
            }
        }
        doencas = new String[saida.length + 1];
        for(int k=0;k<saida.length+1;k++){
            if(k==0){
                doencas[k] = doencagerada[doencagerada.length -1];
            }
            else{
                doencas[k] = saida[k-1]; 
            }
        }
        
        return doencas;
    }
}
```

```java
import java.util.ArrayList;

public class Player {
    int score;
    String name;
    public Player(int score, String name){
        this.score = score;
        this.name = name;
    }

    public int getScore(){
        return score;
    }

    public String getName(){
        return name;
    }

    public void setScore(int score){
        this.score = score;
    }

    public void setName(String name){
        this.name = name;
    }
}
```

```java
public class Leaderboard{
    ArrayList<Player> vencedores;

    public Leaderboard(){
        vencedores = new ArrayList<Player>();
        for(int i=0; i<5; i++){
            vencedores.add(new Player(0, "-"));
        }
    }

    public void addPlayer(int score, String name){
        vencedores.add(new Player(score, name));
        atualizaLeader();
        deletaLeader();
    }

    public void deletaLeader(){
        vencedores.remove(5);
    }

    public void atualizaLeader(){
        int pos_novo = 5;
        for(int i=4; i>=0; i--){
            if(vencedores.get(pos_novo).getScore() > vencedores.get(i).getScore() || vencedores.get(i).getName().equals("-")){
                chanceLeader(pos_novo, i);
                pos_novo--;
            }
        }
    }

    public void chanceLeader(int p1, int p2){
        int p1score = vencedores.get(p1).getScore();
        String p1name = vencedores.get(p1).getName();
        vencedores.get(p1).setScore(vencedores.get(p2).getScore());
        vencedores.get(p1).setName(vencedores.get(p2).getName());
        vencedores.get(p2).setScore(p1score);
        vencedores.get(p2).setName(p1name);
    }

    public void showLeaders(){
        System.out.println("----Podio----");
        for(int i=0; i<5; i++){
            System.out.println("Jogador n "+ i + " Nome: " + vencedores.get(i).getName() + " Pontuação: " + vencedores.get(i).getScore());
        }
    }
}
```
